```python
#3. Write a Pandas program to add, subtract, multiple and divide two Pandas Series. Go to the editor
sample_one = pd.Series([2, 4, 6, 8, 10])
sample_two = pd.Series([1, 3, 5, 7, 9])
sample_one + sample_two
sample_one - sample_two
sample_one/sample_two
sample_one*sample_two
```

```python
#8. Write a Pandas program to convert the first column of a DataFrame as a Series. Go to the editor
dict = {'col1' :[1,2,3,4,7,11], 'col2': [4,5,6,9,5,0], 'col3': [7,5,8,12,1,11]}
df = pd.DataFrame(dict)
col = df.iloc[:, 0] #first arguement is rows, second is column
print(col)
```

```python
# 10. Write a Pandas program to convert Series of lists to one Series. Go to the editor
# Sample Output:
# Original Series of list
# 0 [Red, Green, White]
# 1 [Red, Black]
# 2 [Yellow]
series = pd.Series([['red', 'green', 'white'], ['red', 'black'], ['yellow']])
series = series.apply(pd.Series).stack().reset_index(drop = True)
series
```

```python
#11. Write a Pandas program to sort a given Series. Go to the editor
series = pd.Series(['100', '200', 'python', '300.12', '400'])
pd.Series(series).sort_values(ascending = False)
```

```python
#12. Write a Pandas program to add some data to an existing Series.
series = pd.Series(['100', '200', 'python', '300.12', '400'])
pd.Series(series).append(series)
```
